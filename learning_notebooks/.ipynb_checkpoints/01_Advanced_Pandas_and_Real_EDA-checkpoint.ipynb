```python
# handling missing values 
# detect missing values 
df.isnull()
```

```python
# count missing values per ccolumn
df.isnull().sum()
```

```python
# do not ddrop missing values because data loss 
# df.dropna()
```

```python
# median is better 
df['Loan_Amount'] = df['Loan_Amount'].fillna(df['Loan_Amount'].mean())
```

```python
# catagorical data 
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# advanced methods ffill and bfill
df=df.bfill()
```

```python
df.isnull().sum()
```

```python
# feature engineering 
df['TotalIncome']=df['ApplicantIncome']+df['CoapplicantIncome']
```

```python
df['LoanIncomeRatio']=df['Loan_Amount']/df['TotalIncome']
```

```python
df.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
df.drop(['Property_Area'],axis=1,inplace=True)
```

```python
df
```

```python
# binning 
df['IncomeGroup']=pd.cut(df['TotalIncome'],bins=[0,1000,3000,5000,10000],labels=['low','medium','high','veryhigh'])
```

```python
df
```

```python
# Encoding catogorical data 
# label encoding 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder() 
df['Gender']=le.fit_transform(df['Gender'])
```

```python
df
```

```python
# one hot encoding 
df=pd.get_dummies(df,drop_first=True).astype(int)
```
