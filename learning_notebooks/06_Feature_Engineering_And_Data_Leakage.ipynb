```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Loan_Status']=le.fit_transform(df['Loan_Status'])
```

```python
X=df[['CoapplicantIncome','ApplicantIncome','Gender', 'Married', 'Education']]
y=df['Loan_Status']
```

```python
# pipeline for use seperate num,cat column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
# Pipeline 
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

num_feature=['CoapplicantIncome','ApplicantIncome']
cat_feature=['Gender', 'Married', 'Education']

preprocess=ColumnTransformer([
    ('num',StandardScaler(),num_feature),
    ('cat',OneHotEncoder(),cat_feature)
])

model=Pipeline([
    ('preprocess',preprocess),
    ('classifier',LogisticRegression())
])
model.fit(X_train,y_train)
```

```python
# pipeline with randomfrorest
from sklearn.ensemble import RandomForestClassifier
rf_model=Pipeline([
    ('preprocess',preprocess), 
    ('classifier',RandomForestClassifier(
        n_estimators=200, 
        random_state=42
    ))
])
rf_model.fit(X_train,y_train)
```

```python
# gridsearch with pipeline 
from sklearn.model_selection import GridSearchCV
param_grid={
    'classifier__n_estimators':[100,200], 
    'classifier__max_depth':[5,10]
}
grid=GridSearchCV(
    rf_model, 
    param_grid,
    cv=5,
    scoring='f1'
)
grid.fit(X_train,y_train)
```
